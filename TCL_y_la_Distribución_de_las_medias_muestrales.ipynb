```python
import pandas as pd #Manejo y manipulación de datos
import seaborn as sns #Visualización de datos
```

```python
#Muestra aleatoria de binomiales donde nos arroja el numero de exitos (10 exitos)
np.random.binomial(1000,0.6,10)
#1000 veces que lanzamos una moneda
#0.6 probabilidad de que salga cara
```

```python
df =pd.DataFrame()

for i in range(1,51):
  muestra=np.random.binomial(1000,0.6,10)
  col = f'muestra_{i}'
  df[col] = muestra

df.head(5) #Muestra el dataframe
```

```python
df_medias_muestras = pd.DataFrame(df.mean(),columns=["Medias de la muestra"])
df_medias_muestras
```

```python
sns.distplot(df_medias_muestras)
```
