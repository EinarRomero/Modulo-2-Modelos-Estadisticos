## El problema del encuentro

Romeo y Julieta pretenden encontrarse en un lugar epecifico entre $[0,60]$

* Si llega primero Romeo, entonces piensa esperar 10 minutos.
* Si llega primero Julieta, entonces piensa esperar 10 minutos.

Se elige un numero al azar entre $0$ y $60$. Supongamos que el par de numeros se elige sobre el cuadrado $60^2$, de modo que la probabilidad de un evento, sea el area favorable entre el area total.

Sea el evento:

* A: se encuentren a distancia de 10 minutos. Calculemos la probabilidad de A.




```python
import matplotlib.pyplot as plt #graficos en python
import numpy as np
```

```python
#Vamos a generar 100,000 simulaciones
muestra = 100000

#Generacion de tiempos aleatorios
# x: tiempo de llegada de Romeo [0,60]
# y: tiempo de llegada de Julieta [0,60]

#Generamos una muestra de numeros aleatorios entre 0 y 60
x,y = np.random.uniform(0,60,muestra),np.random.uniform(0,60,muestra)

#Condiciones de encuentro y asigancion de colores
#La  diferencia de tiempo es menor o igual a 10 min para ambos
encuentro = np.where( abs(x-y)<=10,'red','blue')

#La diferencia de tiempo es menor a 10 para ambos y la diferencia de tiempo de julieta con romeo es menor a 5
#encuentro = np.where((x-y<10)& (y-x<5),'red','blue')

#La diferencia de tiempo es menor a 10 para ambos y la diferencia de tiempo de romeo con julieta es menor a 5
#encuentro = np.where((y-x<10)& (x-y<5),'red','blue')

#Calculo de la probabilidad del encuentro
k = 0
for elemento in encuentro:
  if elemento == 'red':
    k += 1

print("La probabilidad del evento A: {}".format(k/muestra))
plt.scatter(x,y,c=encuentro,s=1) #Crea un grafico de dispersion
plt.show()

```
